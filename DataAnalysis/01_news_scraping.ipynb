### Daum 뉴스기사 제목 스크래핑하기

```python
import requests
from bs4 import BeautifulSoup

section_dict = {
    "기후/환경": "tech",
    "사회": "society",
    "경제": "economy",
    "정치": "politics",
    "국제": "world",
    "문화": "culture",
    "생활": "life",
    "IT/과학": "tech",
}


def print_news(category):
    section = section_dict[category]
    print(f"https://news.daum.net/{section}")
    
    url = f"https://news.daum.net/{section}"
    req_header = {
        "user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/104.0.0.0 Safari/537.36"
    }
    res = requests.get(url, headers=req_header)
    res.encoding = "utf-8"  # 한글 깨짐 방지
    print(f"{res.status_code}")
    if res.ok:
        html = res.text
        soup = BeautifulSoup(html, "html.parser")
        strong_tag_list = soup.select("a.item_newsheadline2 strong.tit_txt")
        for strong_tag in strong_tag_list:

            title = strong_tag.text.strip()
            link = strong_tag.find_parent("a")["href"]
            print(link)
            print(title)

    else:
        print(f"에러 코드 = {res.status_code}")


print_news("경제")
```

### Nate 뉴스 기사 제목/이미지 스크래핑

```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin
from IPython.display import Image, display

section_dict = {
    "최신뉴스": "0100",
    "정치": "0200",
    "경제": "0300",
    "사회": "0400",
    "세계": "0500",
    "IT/과학": "0600",
}

def print_news(category):
    section = section_dict[category]
    print(f"===== category : {category} https://news.nate.com/recent?mid=n{section} =====")
    url = f"https://news.nate.com/recent?mid=n{section}"
    
    req_header = {
        "user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/104.0.0.0 Safari/537.36"
    }
    res = requests.get(url,headers=req_header)
    res.encoding = "euc-kr"
    
    if res.ok :
        html = res.text
        soup = BeautifulSoup(html, "html.parser")
        mlt_tag_list = soup.select("div.mlt01")
    
        for mlt_tag in mlt_tag_list:
            title_tag = mlt_tag.select_one("a.lt1 h2.tit")
            url_tag = mlt_tag.select_one("a.lt1")
            img_tag =mlt_tag.select_one("a.lt1 span.ib em.mediatype img")
            title = title_tag.text.strip()
            url =urljoin("https:",url_tag["href"])
            
            
            if img_tag and img_tag["src"]:
                img = urljoin("https:",img_tag["src"])
                display(Image(img))
            
            #print(img)    
            print(title)
            print(url)
            
            
            
            
    else:
        print(f"에러 코드 = {res.status_code}")    

print_news("최신뉴스")
```
